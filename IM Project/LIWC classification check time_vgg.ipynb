````python
def lasso_feature_selection(features, target):
    # 訓練 Lasso 模型，alpha 是正則化參數，可根據需求調整
    lasso = Lasso(alpha=0.001)
    lasso.fit(features, target)  # target 是你的目標變量

    # 選擇重要特徵 (Lasso 係數不為 0 的特徵)
    selected_features = [i for i, weight in enumerate(lasso.coef_) if weight != 0]
    selected_weights = lasso.coef_[lasso.coef_ != 0]
    intercept = lasso.intercept_

    print("Lasso 選擇的特徵索引:", selected_features)

    return selected_features
````

````python
def gradient_boosting_feature_selection(features, target):
    # 訓練 Gradient Boosting 模型
    try:
        gbt = GradientBoostingClassifier()  # 或 GradientBoostingRegressor 根據需要選擇
        gbt.fit(features, target)  # target 是你的目標變量
    except:
        gbt = GradientBoostingRegressor()
        gbt.fit(features, target)

    # 提取特徵重要性
    feature_importances = gbt.feature_importances_

    selected_features_gbt = [i for i, importance in enumerate(feature_importances) if importance > 0]
    print("Gradient Boosting Tree 選擇的特徵索引:", selected_features_gbt)

    return selected_features_gbt
````

````python
def chi2_feature_selection(features, target):
    chi2_selector = SelectKBest(chi2, k=10)
    predictions = chi2_selector.fit_transform(features, target)

    # 查看選擇的特徵
    selected_features = chi2_selector.get_support(indices=True)
    print("選擇的特徵索引:", selected_features)

    return selected_features
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, log_loss
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import KFold
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
import torch.optim as optim
````

````python
class FNN(nn.Module):
    def __init__(self, input_dim, hidden_units_1=32, hidden_units_2=16, output_units=1, dropout_prob=0.2):
        super(FNN, self).__init__()
        self.hidden_layer_1 = nn.Linear(input_dim, hidden_units_1)
        self.hidden_layer_2 = nn.Linear(hidden_units_1, hidden_units_2)
        self.output_layer = nn.Linear(hidden_units_2, output_units)

        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

        # Dropout 層
        self.dropout = nn.Dropout(dropout_prob)

    def forward(self, x):
        x = self.relu(self.hidden_layer_1(x))
        x = self.dropout(x)  # Dropout 用於隱藏層1
        x = self.relu(self.hidden_layer_2(x))
        x = self.dropout(x)  # Dropout 用於隱藏層2
        x = self.sigmoid(self.output_layer(x))
        return x
````

````python
class LSTMClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, dropout_prob=0.2):
        super(LSTMClassifier, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        # 定義 LSTM 層
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout_prob)
        # 定義輸出層，Sigmoid 激活函數使輸出在 (0, 1) 範圍，適合二元分類
        self.fc = nn.Linear(hidden_size, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # 初始化 LSTM 隱藏層和記憶單元
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        # 前向傳播 LSTM
        out, _ = self.lstm(x, (h0, c0))

        # 取最後一層隱藏狀態的輸出並通過 Sigmoid 層
        out = self.fc(out[:, -1, :])
        out = self.sigmoid(out)
        return out
````

````python
class GRUClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, dropout_prob=0.2):
        super(GRUClassifier, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        # 定義 GRU 層
        self.gru = nn.GRU(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout_prob)
        # 定義輸出層，Sigmoid 激活函數使輸出在 (0, 1) 範圍，適合二元分類
        self.fc = nn.Linear(hidden_size, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # 初始化 GRU 隱藏層
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        # 前向傳播 GRU
        out, _ = self.gru(x, h0)

        # 取最後一層隱藏狀態的輸出並通過 Sigmoid 層
        out = self.fc(out[:, -1, :])
        out = self.sigmoid(out)
        return out
````

````python
import time
from tqdm import tqdm

def train_binary_classification_model(model, train_loader, criterion, optimizer, num_epochs, device):
    model.train()
    start_time = time.time()

    for epoch in range(num_epochs):
        epoch_start_time = time.time()  # Time for the current epoch

        running_loss = 0.0
        total_loss = 0
        for i, (inputs, targets) in enumerate(tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}", ncols=100, leave=False)):
        # for inputs, targets in train_loader:
            inputs, targets = inputs.to(device), targets.to(device)

            # 清除優化器的梯度
            optimizer.zero_grad()

            # 前向傳播
            outputs = model(inputs).view(-1)
            loss = criterion(outputs, targets)
            total_loss += loss.item()

            # 反向傳播和優化
            loss.backward()
            optimizer.step()

        epoch_end_time = time.time()
        epoch_duration = epoch_end_time - epoch_start_time
        total_time = epoch_end_time - start_time

        remaining_time = (num_epochs - (epoch + 1)) * (total_time / (epoch + 1))

        avg_loss = total_loss / len(train_loader)

        print(f"\rEpoch [{epoch+1}/{num_epochs}], Loss: {running_loss / len(train_loader):.4f} | "
            f"Epoch Duration: {epoch_duration:.2f} seconds | "
            f"Remaining Time: {remaining_time / 60:.2f} minutes", end="")
        print()
````

````python
def evaluate_binary_classification_model(model, test_loader, device):
    model.eval()  # 切換到評估模式
    all_targets = []
    all_predictions = []

    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs, targets = inputs.to(device), targets.to(device)

            # 預測
            outputs = model(inputs).view(-1)
            predictions = (outputs > 0.5).float()  # 將預測的機率轉為 0 或 1

            # 收集所有目標和預測
            all_targets.extend(targets.cpu().numpy())
            all_predictions.extend(predictions.cpu().numpy())

    # 計算指標
    accuracy = accuracy_score(all_targets, all_predictions)
    precision = precision_score(all_targets, all_predictions)
    recall = recall_score(all_targets, all_predictions)
    f1 = f1_score(all_targets, all_predictions)

    return [accuracy, precision, recall, f1]
````

````python
def evaluate_model_classification(model, X_test, y_test, feature_num):
    y_pred_prob = model.predict(X_test)
    try:
        y_pred_prob = model.predict_proba(X_test)[:, 1]
    except:
        pass

    y_pred = np.where(y_pred_prob > 0.5, 1, 0)
    # 計算評估指標
    accuracy = accuracy_score(y_test[:, feature_num], y_pred)
    precision = precision_score(y_test[:, feature_num], y_pred)
    recall = recall_score(y_test[:, feature_num], y_pred)
    f1 = f1_score(y_test[:, feature_num], y_pred)
    # loss = log_loss(y_test[:, 0], y_pred_prob)

    return [accuracy, precision, recall, f1]
````

````python
def classification(X, y, feature_num):
    kf = KFold(n_splits=10, shuffle=True, random_state=42)
    logistic_score = []
    svm_score = []
    rf_score = []
    xgb_score = []
    nb_score = []
    bnb_score = []
    mnb_score = []
    knn_score = []
    FNN_score = []
    lstm_score = []
    gru_score = []
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # selected_feature = liwc_correlation_feature_selection(feature_num)
        # selected_feature = chi2_feature_selection(X_train, y_train[:, feature_num])
        # selected_feature = lasso_feature_selection(X_train, y_train[:, feature_num])
        # selected_feature = gradient_boosting_feature_selection(X_train, y_train[:, feature_num])
        # X_train, X_test = X_train[:, selected_feature], X_test[:, selected_feature]

        logistic_model = LogisticRegression(solver='newton-cg', max_iter=1000, random_state=42)
        logistic_model.fit(X_train, y_train[:, feature_num])
        logistic_score.append(evaluate_model_classification(logistic_model, X_test, y_test, feature_num))

        svm_model = SVC(kernel='poly', probability=True)  # 如果需要概率輸出，設定 `probability=True`
        svm_model.fit(X_train, y_train[:, feature_num])
        svm_score.append(evaluate_model_classification(svm_model, X_test, y_test,feature_num))

        rf_model = RandomForestClassifier(n_estimators=1000, random_state=42)  # n_estimators 為樹的數量
        rf_model.fit(X_train, y_train[:, feature_num])
        rf_score.append(evaluate_model_classification(svm_model, X_test, y_test, feature_num))

        dtrain = xgb.DMatrix(X_train, label=y_train[:, feature_num])
        dtest = xgb.DMatrix(X_test, label=y_test[:, feature_num])

        # 設定 XGBoost 的參數
        params = {
            'objective': 'binary:logistic',  # 二元分類
            'eval_metric': 'logloss',        # 損失函數
            'max_depth': 3,                  # 樹的最大深度
            'eta': 0.2,                      # 學習率
            'seed': 42                       # 隨機種子
        }

        # 訓練模型
        num_round = 200  # 訓練的迭代次數
        bst = xgb.train(params, dtrain, num_round)
        xgb_score.append(evaluate_model_classification(bst, dtest, y_test, feature_num))

        nb_model = GaussianNB()
        nb_model.fit(X_train, y_train[:, feature_num])
        nb_score.append(evaluate_model_classification(nb_model, X_test, y_test, feature_num))

        bnb_model = BernoulliNB()
        bnb_model.fit(X_train, y_train[:, feature_num])
        bnb_score.append(evaluate_model_classification(bnb_model, X_test, y_test, feature_num))

        mnb_model = MultinomialNB()
        mnb_model.fit(X_train, y_train[:, feature_num])
        mnb_score.append(evaluate_model_classification(mnb_model, X_test, y_test, feature_num))

        knn_model = KNeighborsClassifier(n_neighbors=23)
        knn_model.fit(X_train, y_train[:, feature_num])
        knn_score.append(evaluate_model_classification(knn_model, X_test, y_test, feature_num))

        input_size = X_train.shape[1]   # 輸入特徵維度
        hidden_size = 16 # 隱藏層單元數
        num_layers = 5   # LSTM 層數
        output_units = 1
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

        X_train = torch.tensor(X_train, dtype=torch.float32)
        y_train = torch.tensor(y_train[:, feature_num], dtype=torch.float32)
        X_test = torch.tensor(X_test, dtype=torch.float32)
        y_test = torch.tensor(y_test[:, feature_num], dtype=torch.float32)

        X_train_lstm = X_train.unsqueeze(1)
        X_test_lstm = X_test.unsqueeze(1)

        train_dataset = TensorDataset(X_train, y_train)
        test_dataset = TensorDataset(X_test, y_test)
        train_dataset_lstm = TensorDataset(X_train_lstm, y_train)
        test_dataset_lstm = TensorDataset(X_test_lstm, y_test)

        train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
        test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
        train_loader_lstm = DataLoader(train_dataset_lstm, batch_size=16, shuffle=True)
        test_loader_lstm = DataLoader(test_dataset_lstm, batch_size=16, shuffle=False)

        fnn_model = FNN(input_dim=input_size, output_units=output_units).to(device)
        lstm_model = LSTMClassifier(input_size, hidden_size, num_layers).to(device)
        gru_model = GRUClassifier(input_size, hidden_size, num_layers).to(device)


        criterion = nn.BCELoss()  # 使用二元交叉熵損失
        fnn_optimizer = optim.Adam(fnn_model.parameters(), lr=0.001)
        lstm_optimizer = optim.Adam(lstm_model.parameters(), lr=0.001)  # 使用 Adam 優化器
        gru_optimizer = optim.Adam(gru_model.parameters(), lr=0.001)  # 使用 Adam 優化器
        num_epochs = 100

        # 訓練模型
        train_binary_classification_model(fnn_model, train_loader, criterion, fnn_optimizer, num_epochs, device)
        train_binary_classification_model(lstm_model, train_loader_lstm, criterion, lstm_optimizer, num_epochs, device)
        train_binary_classification_model(gru_model, train_loader_lstm, criterion, gru_optimizer, num_epochs, device)

        # 測試模型
        # 假設 model 和 test_loader 已定義
        FNN_score.append(evaluate_binary_classification_model(fnn_model, test_loader, device))
        lstm_score.append(evaluate_binary_classification_model(lstm_model, test_loader_lstm, device))
        gru_score.append(evaluate_binary_classification_model(gru_model, test_loader_lstm, device))

    logistic_score = np.mean(np.array(logistic_score), axis=0)
    svm_score = np.mean(np.array(svm_score), axis=0)
    rf_score = np.mean(np.array(rf_score), axis=0)
    xgb_score = np.mean(np.array(xgb_score), axis=0)
    nb_score = np.mean(np.array(nb_score), axis=0)
    bnb_score = np.mean(np.array(bnb_score), axis=0)
    mnb_score = np.mean(np.array(mnb_score), axis=0)
    knn_score = np.mean(np.array(knn_score), axis=0)
    FNN_score = np.mean(np.array(FNN_score), axis=0)
    lstm_score = np.mean(np.array(lstm_score), axis=0)
    gru_score = np.mean(np.array(gru_score), axis=0)

    classification_result = pd.DataFrame([logistic_score, svm_score, rf_score, xgb_score, nb_score, bnb_score, mnb_score, knn_score, FNN_score, lstm_score, gru_score], columns = ['accuracy', 'precision', 'recall', 'f1'])
    display(classification_result)
````

````python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import BayesianRidge
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# 定義一個函數來打印評估指標
def evaluate_model(model, X_test, y_test):
    predictions = model.predict(X_test)
    mse = mean_squared_error(y_test, predictions)
    mae = mean_absolute_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)

    return [mse, mae, r2]


def regression(X, y, feature_num):
    kf = KFold(n_splits=10, shuffle=True, random_state=42)

    linear_score = []
    ridge_score = []
    lasso_score = []
    svr_score = []
    tree_score = []
    rf_score = []
    gbr_score = []
    knn_score = []
    bayesian_score = []
    nn_score = []
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # selected_feature = liwc_correlation_feature_selection(feature_num)
        # selected_feature = chi2_feature_selection(X_train, y_train[:, feature_num])
        # selected_feature = gradient_boosting_feature_selection(X_train, y_train[:, feature_num])

        # X_train, X_test = X_train[:, selected_feature], X_test[:, selected_feature]

        linear_model = LinearRegression()
        linear_model.fit(X_train, y_train[:, feature_num])
        linear_score.append(evaluate_model(linear_model, X_test, y_test[:, feature_num]))

        # Ridge 和 Lasso 回歸
        ridge_model = Ridge(alpha=1.0)
        ridge_model.fit(X_train, y_train[:, feature_num])
        ridge_score.append(evaluate_model(ridge_model, X_test, y_test[:, feature_num]))

        lasso_model = Lasso(alpha=0.1)
        lasso_model.fit(X_train, y_train[:, feature_num])
        lasso_score.append(evaluate_model(lasso_model, X_test, y_test[:, feature_num]))

        # 支持向量回歸
        svr_model = SVR(kernel='rbf')
        svr_model.fit(X_train, y_train[:, feature_num])
        svr_score.append(evaluate_model(svr_model, X_test, y_test[:, feature_num]))

        # 決策樹回歸
        tree_model = DecisionTreeRegressor()
        tree_model.fit(X_train, y_train[:, feature_num])
        tree_score.append(evaluate_model(tree_model, X_test, y_test[:, feature_num]))

        # 隨機森林回歸
        rf_model = RandomForestRegressor(n_estimators=100)
        rf_model.fit(X_train, y_train[:, feature_num])
        rf_score.append(evaluate_model(rf_model, X_test, y_test[:, feature_num]))

        # 梯度提升回歸
        gbr_model = GradientBoostingRegressor(n_estimators=100)
        gbr_model.fit(X_train, y_train[:, feature_num])
        gbr_score.append(evaluate_model(gbr_model, X_test, y_test[:, feature_num]))

        # KNN 回歸
        knn_model = KNeighborsRegressor(n_neighbors=23)
        knn_model.fit(X_train, y_train[:, feature_num])
        knn_score.append(evaluate_model(knn_model, X_test, y_test[:, feature_num]))

        # 貝葉斯回歸
        bayesian_model = BayesianRidge()
        bayesian_model.fit(X_train, y_train[:, feature_num])
        bayesian_score.append(evaluate_model(bayesian_model, X_test, y_test[:, feature_num]))

        # 神經網絡回歸
        nn_model = MLPRegressor(hidden_layer_sizes=(100,), max_iter=1000)
        nn_model.fit(X_train, y_train[:, feature_num])
        nn_score.append(evaluate_model(nn_model, X_test, y_test[:, feature_num]))

    linear_score = np.mean(np.array(linear_score), axis=0)
    ridge_score = np.mean(np.array(ridge_score), axis=0)
    lasso_score = np.mean(np.array(lasso_score), axis=0)
    svr_score = np.mean(np.array(svr_score), axis=0)
    tree_score = np.mean(np.array(tree_score), axis=0)
    rf_score = np.mean(np.array(rf_score), axis=0)
    gbr_score = np.mean(np.array(gbr_score), axis=0)
    knn_score = np.mean(np.array(knn_score), axis=0)
    bayesian_score = np.mean(np.array(bayesian_score), axis=0)
    nn_score = np.mean(np.array(nn_score), axis=0)

    regression_result = pd.DataFrame([linear_score, ridge_score, lasso_score, svr_score, tree_score, rf_score, gbr_score, knn_score, bayesian_score, nn_score], columns = ['mse', 'mae', 'r2'])
    display(regression_result)
````

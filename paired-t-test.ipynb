```python
import pandas as pd
from scipy import stats

# Assuming data1 and data2 are your two DataFrames or NumPy arrays with shape (60, 91)
# Example of creating dummy data
data1 = df_ig
data2 = df_threads

# Significance level
alpha = 0.05

# Store the results of t-tests for each column
t_test_results = {'Column': [], 't-statistic': [], 'p-value': [], 'Significant': []}

# Perform a paired t-test for each column
for col in data1.columns:
    # Calculate t-statistic and p-value
    t_stat, p_value = stats.ttest_rel(data1[col], data2[col])

    # Calculate degrees of freedom (number of pairs - 1)
    n = len(data1[col])
    df = n - 1

    # Calculate the critical t-value for the given alpha and degrees of freedom
    critical_value = stats.t.ppf(1 - alpha / 2, df)  # two-tailed test

    # Determine if the result is significant based on the critical t-value
    is_significant = abs(t_stat) > critical_value

    # Store results in dictionary
    if is_significant != True:
        t_stat = None
        p_value = None
    t_test_results['Column'].append(col)
    t_test_results['t-statistic'].append(t_stat)
    t_test_results['p-value'].append(p_value)
    t_test_results['Significant'].append(is_significant)

# Convert results to a DataFrame for easy viewing
t_test_results_df = pd.DataFrame(t_test_results)

# Display the results
print(t_test_results_df)
```
